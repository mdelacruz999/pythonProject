# Covid DataSet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('./covid-data.csv')
```

```python

data_frame.head()
```

```python
# Separo los datos de colombia
data_colombia = data_frame[data_frame['location'].str.contains('Colombia')]
data_colombia.reset_index(drop=True)
```

```python
# Inicio el analisis con los datos del aÃ±o 2020
date_2020 = data_colombia[data_colombia['date'].str.contains('2020')]
date_2020.reset_index(drop=True)
casos_2020 = date_2020['new_cases'].sum()
casos_2020
```

```python
casos_por_mes_2020 = []
muertes_por_mes_2020 = []
```

```python
# Casos - Muertes marzo 2020
marzo_20 = date_2020[date_2020['date'].str.contains('03-2020')]
casos_por_mes_2020.append(marzo_20['new_cases'].sum())
muertes_por_mes_2020.append(marzo_20['new_deaths'].sum())
```

```python
# Casos - Muertes abril 2020
abril_20 = date_2020[date_2020['date'].str.contains('04-2020')]
casos_por_mes_2020.append(abril_20['new_cases'].sum())
muertes_por_mes_2020.append(abril_20['new_deaths'].sum())
```

```python
# Casos - Muertes mayo 2020
mayo_20 = date_2020[date_2020['date'].str.contains('05-2020')]
casos_por_mes_2020.append(mayo_20['new_cases'].sum())
muertes_por_mes_2020.append(mayo_20['new_deaths'].sum())
```

```python
# Casos - Muertes junio 2020
junio_20 = date_2020[date_2020['date'].str.contains('06-2020')]
casos_por_mes_2020.append(junio_20['new_cases'].sum())
muertes_por_mes_2020.append(junio_20['new_deaths'].sum())
```

```python
# Casos - Muertes julio 2020
julio_20 = date_2020[date_2020['date'].str.contains('07-2020')]
casos_por_mes_2020.append(julio_20['new_cases'].sum())
muertes_por_mes_2020.append(julio_20['new_deaths'].sum())
```

```python
# Casos - Muertes agosto 2020
agosto_20 = date_2020[date_2020['date'].str.contains('08-2020')]
casos_por_mes_2020.append(agosto_20['new_cases'].sum())
muertes_por_mes_2020.append(agosto_20['new_deaths'].sum())
```

```python
# Casos - Muertes septiembre 2020
septiembre_20 = date_2020[date_2020['date'].str.contains('09-2020')]
casos_por_mes_2020.append(septiembre_20['new_cases'].sum())
muertes_por_mes_2020.append(septiembre_20['new_deaths'].sum())
```

```python
# Casos - Muertes octubre 2020
octubre_20 = date_2020[date_2020['date'].str.contains('10-2020')]
casos_por_mes_2020.append(octubre_20['new_cases'].sum())
muertes_por_mes_2020.append(octubre_20['new_deaths'].sum())
```

```python
# Casos - Muertes noviembre 2020
noviembre_20 = date_2020[date_2020['date'].str.contains('11-2020')]
casos_por_mes_2020.append(noviembre_20['new_cases'].sum())
muertes_por_mes_2020.append(noviembre_20['new_deaths'].sum())
```

```python
# Casos - Muertes diciembre 2020
diciembre_20 = date_2020[date_2020['date'].str.contains('12-2020')]
casos_por_mes_2020.append(diciembre_20['new_cases'].sum())
muertes_por_mes_2020.append(diciembre_20['new_deaths'].sum())
```

```python
casos_por_mes_2020, muertes_por_mes_2020
```

```python
meses_2020= ['Mar','Abr','May','Jun','Jul','Ago', 'Sep','Oct','Nov','Dic']
```

```python
#Depuracion de informaciÃ³n por casos y muertes respecto a meses 2020
casos_muertes_mes = pd.DataFrame(list(zip(meses_2020,casos_por_mes_2020, muertes_por_mes_2020)), columns = ['Meses','Casos', 'Muertes'])
print(casos_muertes_mes)
```

```python
plt.bar(meses_2020,casos_por_mes_2020)
plt.xlabel("Meses_2020",fontdict=None)
plt.ylabel("Casos Covid 2020",fontdict=None)

plt.show()
```

```python
# Se comprueba que las muertes coincidan con la suma de toda la lista
muertes_2020 = date_2020['new_deaths'].sum()
muertes_2020
print(sum(muertes_por_mes_2020), muertes_2020)
```

```python
# Calculamos las muertes por numero de casos registrados
muertes_por_casos = np.around((muertes_2020 / casos_2020) * 1000, 3)
muertes_por_casos
```

```python
# Transfomo la str a date
date_2020['date'] = pd.to_datetime(date_2020['date'], infer_datetime_format=True)
date_2020_newindex = date_2020.reset_index(drop=True)
```

```python
date_2020_newindex
```

```python
# Calculo los dias desde que se reporto el primer caso, hasta el fin de aÃ±o
date_2020_newindex['date'][300] - date_2020_newindex['date'][0]
```

```python
# Calculamos los casos por dia en el aÃ±o 2020
casos_por_dia_2020 = np.divide(casos_2020, 211)
np.around(casos_por_dia_2020)
```
